```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
Test = pd.read_csv("Test.csv")
Test = pd.DataFrame(Test)
Test
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer


num_encoder = Pipeline([
    ("imputer",SimpleImputer()),
    ("scaler",StandardScaler())
])

text_encoder = OneHotEncoder()

def data_preprocessing(Data):
    a = 'Survived'
    Return_Label = False
    Test_Data = False
    if a in Data.columns:
        label = Data["Survived"]
        Feature = Data.drop(["PassengerId","Survived","Name","Ticket","Cabin"],axis=1)
        Return_Label = True
    else:
        Feature = Data.drop(["PassengerId","Name","Ticket","Cabin"],axis=1)
        Test_Data = True
    
    print(Feature)
    text_data = Feature[["Sex","Embarked"]]
    num_data = Feature.drop(["Sex","Embarked"],axis=1) 
    
    
    
    if Test_Data:
        num_prepared = num_encoder.transform(num_data)
        text_prepared = text_encoder.transform(text_data)
    else:
        num_prepared = num_encoder.fit_transform(num_data)
        text_prepared = text_encoder.fit_transform(text_data)
    
    full_data = np.c_[num_prepared,text_prepared.toarray()]
    print(full_data)
    #print(full_data[0])
    #print("------------------")
    #print(pd.DataFrame(num_prepared,columns=["PassengerId","Pclass","Age","SibSp","Parch","Fare"]))
    #print("------------------")
    #print(text_prepared[200:350])
    
    if Return_Label:
        return full_data,label
    else:
        return full_data
```

```python
Training_Data,Training_label = data_preprocessing(Training)
```

```python
Test_data = data_preprocessing(Test)
```

### Test a Model

```python
from sklearn.ensemble import RandomForestClassifier

RnR = RandomForestClassifier()
RnR = RnR.fit(Training_Data,Training_label)
```

```python
RnR.predict(Training_Data)
```

```python
RnR.predict(Test_data)
```

### Evaluating the Error

```python
from sklearn.metrics import mean_squared_error

predictions = RnR.predict(Training_Data)
mse = mean_squared_error(predictions,Training_label)
rmse = np.sqrt(mse)
print(rmse)
```

### Try another Model

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR = LR.fit(Training_Data,Training_label)
predictions = LR.predict(Training_Data)
mse = mean_squared_error(predictions,Training_label)
rmse = np.sqrt(mse)
print(rmse)
```

### Cross_Validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(RnR,Training_Data,Training_label,cv=10)
rmse_scores = np.sqrt(scores)
scores
```

```python
def display_scores(scores):
    print("Scores:",scores)
    print("Mean:",scores.mean())
    print("Standard Deviation:",scores.std())
display_scores(scores)
```

```python
scores2 = cross_val_score(LR,Training_Data,Training_label,cv=10)
rmse_scores2 = np.sqrt(scores)
scores2
```
