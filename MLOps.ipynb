```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
Training = pd.read_csv("train.csv")
Training = pd.DataFrame(Training)
Training
```

```python
Test = pd.read_csv("Test.csv")
Test = pd.DataFrame(Test)
Test
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer


num_encoder = Pipeline([
    ("imputer",SimpleImputer()),
    ("scaler",StandardScaler())
])

text_encoder = OneHotEncoder()

def data_preprocessing(Data):
    a = 'Survived'
    Return_Label = False
    Test_Data = False
    if a in Data.columns:
        label = Data["Survived"]
        Feature = Data.drop(["PassengerId","Survived","Name","Ticket","Cabin"],axis=1)
        Return_Label = True
    else:
        Feature = Data.drop(["PassengerId","Name","Ticket","Cabin"],axis=1)
        Test_Data = True
    
    print(Feature)
    text_data = Feature[["Sex","Embarked"]]
    num_data = Feature.drop(["Sex","Embarked"],axis=1) 
    
    
    
    if Test_Data:
        num_prepared = num_encoder.transform(num_data)
        text_prepared = text_encoder.transform(text_data)
    else:
        num_prepared = num_encoder.fit_transform(num_data)
        text_prepared = text_encoder.fit_transform(text_data)
    
    full_data = np.c_[num_prepared,text_prepared.toarray()]
    
    if Return_Label:
        return full_data,label
    else:
        return full_data
```

```python
Training_Data,Training_label = data_preprocessing(Training)
```

```python
Test_data = data_preprocessing(Test)
```

```python
model = RandomForestClassifier( n_estimators=100,    #default=100
                                max_depth=6,         #default=None
                                min_samples_split=8, #default=2,
                                min_samples_leaf=1,  #default=1,
                                bootstrap=True,      #default=True
                                ccp_alpha=0.0,       #default=0.0
                                random_state=None)
```

```python
model.fit(Training_Data,Training_label)
```

```python
F_Result = model.predict(Test_data)
```
